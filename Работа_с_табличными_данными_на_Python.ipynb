```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
output_notebook()
```

```
Table = pd.read_excel('https://drive.google.com/uc?export=download&id=1nEovvreO_-BHd5ebl2O-kVj_AaLfnHci')
Table.head()
```

```
Table.tail()
```

```
print('Расположение месяцев в порядке возрастания выработки')
Table1 = Table.sort_values(['electricity generation'])
print(Table1)
print()
```

```
print('Расположение месяцев в порядке убывания выработки')
Table2 = Table.sort_values(['electricity generation'], ascending = False)
print(Table2)
print()
```

```
Table3 = Table.copy()
Table3['aopd'] = Table3['electricity generation']/30 #aopd- approximate output per day(примерная средня выработка за сутки в этом месяце)
Table3
```

```
print('месяцы в которых суточная выработка больше 3000') 
print(Table3[Table3['aopd'] > 3000.]) 
print() 
```

```
print(' Данные по выработке за 2013г ')
print(Table3.iloc[:-12 , :])
```

```

Table3['Mounth'] = np.where(Table3['Date'].str.find('01.01') >= 0, 'Январь', np.where(Table3['Date'].str.find('01.02') >= 0, 'Февраль', np.where(Table3['Date'].str.find('01.03') >= 0, 'Март', np.where(Table3['Date'].str.find('01.04') >= 0, 'Апрель', np.where(Table3['Date'].str.find('01.05') >= 0, 'Май', np.where(Table3['Date'].str.find('01.06') >= 0, 'Июнь', np.where(Table3['Date'].str.find('01.07') >= 0, 'Июль', np.where(Table3['Date'].str.find('01.08') >= 0, 'Август', np.where(Table3['Date'].str.find('01.09') >= 0, 'Сентябрь', np.where(Table3['Date'].str.find('01.10') >= 0, 'Октябрь', np.where(Table3['Date'].str.find('01.11') >= 0, 'Ноябрь', np.where(Table3['Date'].str.find('01.12') >= 0, 'Декабрь','?'))))))))))))
Table3
```

```
Table3['Year'] = np.where(Table3['Date'].str.find('2013') >= 0, '2013', np.where(Table3['Date'].str.find('2014') >= 0, '2014', '?'))
Table3['Year']=Table3['Year'].astype(int)
Table3
```

```

print('Максимальная среднесуточная выработка по месяцам')
f=Table3.groupby(['Mounth'])['aopd'].max()
print(f)
print('Суммарная среднесуточная выработка по месяцам')
f2=Table3.groupby(['Mounth'])['aopd'].sum()
print(f2)
```

```
f.plot(kind = 'bar', x='Mounth', y='aopd')
print()
```

```
generation_sum = f2
labels = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь', 'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']

fig1, ax1 = plt.subplots(figsize = (10, 10))
ax1.pie(generation_sum, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```

p = figure(plot_width = 800, plot_height = 400)

colors = ['magenta', 'red']

for Year in range(2013, 2015):
  
  values = Table3[Table3['Year'] == Year]['aopd'].values 
  
  p.line(np.arange(12), values, color = colors[Year - 2013], legend_label = str(Year), line_width = 3)

  p.circle(np.arange(12), values, color = colors[Year - 2013], alpha=0.5, size = 8)


show(p)
```

```
input_matrix = input_df.values # перевод таблицы в матрицу
print('размерность матрицы', input_matrix.shape)
print()
for i in range(1, len(input_df.columns)):
  print("{: <12} mean = {:.2f}".format(input_df.columns[i], input_matrix[ : , i].mean()))
```

```
fig, ax = plt.subplots(8, 1, figsize = (15, 12))

timestamps = range(0, input_matrix.shape[0], 5)

# зададим диапазон
begin = 0
end   = 141 
step = 5
ax[0].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 3], 'black', marker = '.', linewidth = 3, markersize = 9)
ax[1].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 4], 'red', linestyle = '--')
ax[2].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 5], 'gray', linestyle = '-')
ax[3].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 6], 'green', marker = 'v')
ax[4].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 7], 'Blue', marker = 'x')
ax[5].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 8], 'purple', marker = '*')
ax[6].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 9], 'yellow', marker = 'o')
ax[7].plot(timestamps[begin : end : step], input_matrix[begin : end : step, 10], 'brown', marker = '+')

legend = ['уголь ◼', 'ТЭЦ ', 'атом ☢', 'ветер ↠', 'ГЭС ≈', 'биомасса', 'солнышко ☼', 'ocgt']

for _ in zip(ax, legend):
  _[0].set_ylabel('Мощность, ГВт')
  _[0].legend([_[1]])
  _[0].grid(True)
ax[-1].set_xlabel('Время, мин')

plt.show()

#не понятно как сделать то же для нашей работы
```
